### Conditional Statements (`if`, `elif`, and `else`):

- Conditional statements allow the execution of different blocks of code based on whether a condition is true or false.


```python
x = 10
if x > 5:
    print("x is greater than 5")
elif x == 5:
    print("x is equal to 5")
else:
    print("x is less than 5")
```

```python
if x > 12:
    print('x is greater then 5')
```

```python
if x > 12:
    print('x is greater then 12')
else:
    print('X is less then 12')
```

### Nested if Statement
- if statement can also be checked inside other if statement. This conditional statement is called a nested if statement. 
- This means that inner if condition will be checked only if outer if condition is true and by this, we can see multiple conditions to be satisfied.

```python
Image('https://media.geeksforgeeks.org/wp-content/uploads/nested-if.jpg')
```

```python
# Nested if statement example
num = 10

if num > 5:
    print("Bigger than 5")

if num <= 15:
    print("Between 5 and 15")
```

### IF-Elif

- if-elif StatementThe if-elif statement is shortcut of if..else chain. 
- While using if-elif statement at the end else block is added which is performed if none of the above if-elif statement is true.

```python
Image('https://www.programsbuzz.com/sites/default/files/inline-images/Screenshot%202021-06-07%20at%201.08.41%20PM.png')
```

```python
# if-elif statement example

letter = "A"

if letter == "B":
	print("letter is B")

elif letter == "C":
	print("letter is C")

elif letter == "A":
	print("letter is A")

else:
	print("letter isn't A, B or C")
```

### Scenario: Internet Speed Classification
- Imagine an application that tests a user's internet speed and classifies the quality of their connection based on the speed results. 
- The application can then give recommendations to the user based on their connection speed.

```python
# Suppose this is the measured download speed in Mbps
download_speed = int(input('Internet speed')) # You can change this value to test different speeds

if download_speed < 1:
    print("Your internet connection is very slow. You might face difficulties in even basic web browsing.")
elif 1 <= download_speed < 5:
    print("Your internet connection is slow. You can do basic browsing, but streaming videos will be challenging.")
elif 5 <= download_speed < 25:
    print("Your internet connection is average. You can stream videos at HD quality.")
elif 25 <= download_speed < 100:
    print("Your internet connection is fast. You can stream videos at 4K quality and enjoy smooth online gaming.")
else:
    print("Your internet connection is super-fast. You're all set for any online activity!")
```
